```python
import pandas as pd
from pycaret.time_series import *
```

```python
data = pd.read_csv (r'/home/sulyokm/headlousepycaret.csv', delimiter=";")   
df = pd.DataFrame(data, columns= ['Week','google', 'pandemic'])
df['Week'] = pd.to_datetime(df['Week'])
```

```python
s = df.set_index('Week')['google']
s=pd.Series(s)
#print(s)
type(s)
```

```python

setup(s, target = 'google', fh = 12, fold = 3, session_id = 123)
check_stats()
plot_model(plot='diagnostics')
plot_model(plot='acf')
plot_model(plot='cv')
plot_model(plot='ts')
#plot_model(plot='decomp_stl')
# functional API


```

```python
best = compare_models()
```

```python

tuned = tune_model(best)
print(best)
plot_model(best, plot = 'forecast')
plot_model(best, plot = 'residuals')
plot_model(best, plot = 'diagnostics')
plot_model(best, plot = 'insample')
# finalize model
final_best = finalize_model(best)# generate predictions
predict_model(final_best, fh = 90)
# save the model
save_model(final_best, 'my_best_model')

```
